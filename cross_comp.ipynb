```python
import os
import pandas as pd
import numpy as np
```

```python

def concat_csv_files(folder_path):
  # Get all CSV files in the folder
  csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

  # Create an empty DataFrame to store the concatenated data
  df = pd.DataFrame()

  # Iterate over all CSV files and read them into the DataFrame
  for file in csv_files:
    file_path = os.path.join(folder_path, file)
    temp_df = pd.read_csv(file_path)
    df = pd.concat([df, temp_df], ignore_index=True)

  return df

# Example usage
folder_path = 'ps'  # Replace with the actual path to your folder
df = concat_csv_files(folder_path)

# Print the concatenated DataFrame
print(df)
```

```python
new_vals = {}
original_vals = df.values
for tipo, fls, lr, do, bs, j, mae in original_vals:
    if (tipo, fls, lr, do, bs) in new_vals.keys():
        new_vals[(tipo, fls, lr, do, bs)] += mae
    else:
        new_vals[(tipo, fls, lr, do, bs)] = mae
    
for k in new_vals:
    new_vals[k] = new_vals[k]/3
new_vals
```

```python
novo_df = pd.DataFrame(new_vals.values(), index=new_vals.keys())
novo_df = novo_df.sort_values(by=0, ascending=True)
novo_df
```
